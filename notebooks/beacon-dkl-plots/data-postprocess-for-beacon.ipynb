```python
# read files
import pickle

def read_pickle_al(file_path: str) -> dict:
    with open(file_path, "rb") as f:
        pkl_data_al = pickle.load(f)
    return pkl_data_al

```

```python
file_path_al_ei = "beacon-plotting-data/dkl-EI/Active_learning_statistics.pkl"
file_path_al_mu = "beacon-plotting-data/dkl-MU(high-beta-UCB)/Active_learning_statistics.pkl"
file_path_al_bn = "beacon-plotting-data/dkl-bn/Active_learning_statistics.pkl"

pkl_data_al_ei = read_pickle_al(file_path=file_path_al_ei)
pkl_data_al_mu = read_pickle_al(file_path=file_path_al_mu)
pkl_data_al_bn = read_pickle_al(file_path=file_path_al_bn)
```

```python
# 
```

```python
def get_acq_ccords_seed_indices_and_t(pkl_data_al: dict):
    indices_all = np.asarray(pkl_data_al["indices_all"])          # (N, 2)
    acq_idx = np.asarray(pkl_data_al["acquired_order"])           # (T,)
    acq_coords = indices_all[acq_idx]                             # (T, 2)
    t = np.arange(len(acq_coords))
    
    #
    seed_indices = np.asarray(pkl_data_al["seed_indices"])  # Initial sampled indices (referring to positions in indices_all)
    seed_coords = indices_all[seed_indices]
    return t, acq_coords, seed_coords
```

```python
# prep results for 3 cases

## EI
t_ei, acq_coords_e, seed_coords_ei = get_acq_ccords_seed_indices_and_t(pkl_data_al_ei)


## UCB-MU
t_mu, acq_coords_mu, seed_coords_mu  = get_acq_ccords_seed_indices_and_t(pkl_data_al_mu)


## Beacon
t_bn, acq_coords_bn, seed_coords_bn  = get_acq_ccords_seed_indices_and_t(pkl_data_al_bn)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a consistent color palette to match your spatial plots
# Using colors from the 'plasma' or standard high-contrast sets
colors = {
    "EI": "#1f77b4",      # Strong Blue
    "MU": "#e377c2",      # Pinkish (similar to UCB-MU style)
    "Beacon": "#ff7f0e"   # Bright Orange (matches the "Beacon" heat)
}

plt.figure(figsize=(10, 5))
ax = plt.gca()

# Data plotting
data_sets = [
    ("EI", t_ei, pkl_data_al_ei),
    ("MU", t_mu, pkl_data_al_mu),
    ("Beacon", t_bn, pkl_data_al_bn)
]

for label, t, pkl in data_sets:
    y = pkl["mean_y_pred_mean_al"]
    
    # Plot the main line
    line, = ax.plot(t, y, linewidth=2.5, label=label, color=colors[label], alpha=0.9)
    
    # Add a final 'target' marker at the last point for quick visual scanning
    ax.scatter(t[-1], y[-1], color=colors[label], s=60, edgecolors='k', zorder=5)

# Aesthetics
ax.set_xlabel("Step (Iteration)", fontsize=12, fontweight='bold')
ax.set_ylabel("Surrogate Prediction Mean", fontsize=12, fontweight='bold')
ax.set_title("Active Learning Convergence Comparison", fontsize=14, pad=15)

# Improve grid and spines
ax.grid(True, linestyle='--', alpha=0.6)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Legend with a frame for legibility
ax.legend(loc="best", frameon=True, shadow=True, borderpad=1)

plt.tight_layout()
plt.show()
```

```python
## tajectory
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Assuming img, t_ei, acq_coords_e, etc. are already defined in your environment

cases = [
    ("EI",     t_ei, acq_coords_e,  seed_coords_ei),
    ("UCB-MU", t_mu, acq_coords_mu, seed_coords_mu),
    ("Beacon", t_bn, acq_coords_bn, seed_coords_bn),
]

t_max = max(np.max(t) for _, t, _, _ in cases) if cases else 1

# Increase figsize slightly for better spacing
fig, axes = plt.subplots(1, 3, figsize=(20, 7), sharex=True, sharey=True)

for ax, (name, t, acq_xy, seed_xy) in zip(axes, cases):
    # Background image
    ax.imshow(img, cmap="gray", origin="upper", aspect='equal')

    # 1. Improved Path Line: Use a dark color with higher alpha or a contrasting color
    ax.plot(acq_xy[:, 0], acq_xy[:, 1],
            linestyle="-", linewidth=1.2, alpha=0.5, color="cyan")

    # 2. Seeds: Slightly larger for visibility
    ax.scatter(seed_xy[:, 0], seed_xy[:, 1],
               s=60, c="white", edgecolors="black", linewidths=1,
               marker="o", label="Seed", zorder=3)

    # 3. Acquired Points: Using 'plasma' for better visibility on gray
    # Changed marker to 'o' with edges to make colors easier to see
    sc = ax.scatter(acq_xy[:, 0], acq_xy[:, 1],
                    c=t, cmap="plasma", vmin=0, vmax=t_max,
                    s=50, marker="o", edgecolors="k", linewidths=0.5, 
                    label="Acquired", zorder=4)

    # 4. Start/End markers: Larger and more distinct
    ax.scatter(acq_xy[0, 0], acq_xy[0, 1],
               s=200, marker="*", c="gold", edgecolors="k", 
               linewidths=1, label="Start", zorder=5)
    ax.scatter(acq_xy[-1, 0], acq_xy[-1, 1],
               s=200, marker="P", c="lime", edgecolors="k", 
               linewidths=1, label="End", zorder=5)

    ax.set_title(name, fontsize=14, fontweight='bold')
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True, alpha=0.15, color="white")

# Simplified Legend Logic
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=4, 
           frameon=True, facecolor='white', framealpha=0.8, fontsize=12)

# Professional Colorbar Placement
fig.subplots_adjust(right=0.92, bottom=0.15)
cbar_ax = fig.add_axes([0.94, 0.25, 0.015, 0.5]) # [left, bottom, width, height]
cbar = fig.colorbar(sc, cax=cbar_ax)
cbar.set_label("Step Number", fontsize=12, labelpad=10)

plt.show()
```
